```python
import glob
```

```python
files = glob.glob('Data/*.txt')
```

```python
files
```

```python
import re
```

```python
for file in files:
    df = pd.read_csv(file)
    df.columns = [co.lower() for co in df.columns]
    df = df[df.fipstate == 6]
    name = re.findall(r'^(Data/\w*).txt', file)
    new_name = name[0] + '_mod.csv'
    df.to_csv(new_name, header=True)
    print('Just finished file %s' % (name))
```
